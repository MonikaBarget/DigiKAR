```python
import pywikibot
from pywikibot import pagegenerators as pg
import pandas as pd
import csv
```

```python
# STEP2: Running new query to find COUNTRY and TERRITORY

# open csv file with Wikidata IDs

entity_file="https://raw.githubusercontent.com/ieg-dhr/DigiKAR/main/OntologyFiles/Places_Geonames_Wikidata.csv"
df = pd.read_csv(entity_file, index_col=0)
id_list=df['Wikidata'].tolist() 

# define function for country information

def check_for_country(item):
    item_dict = item.get()
    claims = item_dict["claims"]
    try:
        countries = claims["P17"]
        for n in range(len(countries)):
            country_claim=countries[n] # to get latest edit, add [0]
            #print(country_claim)
            country_id=country_claim.target.id
            country_label=country_claim.target.labels["de"]
            print(country_id, country_label)
    except:
        print("No country found.")

# iterate over IDs in list and find information

for i in id_list:
    i_list=[]
    with open('C:\\Users\\p70076654\\Downloads\\SPARQL_Country_Territory.rq', 'r') as query_file:
        QUERY = query_file.read()
            
# create generator based on query
# returns an iterator that produces a sequence of values when iterated over
# useful when creating large sequences of values
    try: 
        wikidata_site = pywikibot.Site("wikidata", "wikidata")
        
# start query using string interpolation        
# https://stackoverflow.com/questions/76762185/matching-specific-geonames-ids-with-wikidata-ids-using-pywikibot
        generator = pg.WikidataSPARQLPageGenerator(QUERY % i, site=wikidata_site)
        #print("Successfully excuted query: ", generator)

    # iterate over generator and print Wikidata ID 
        for item in generator:
            if 'de' in item.labels:
                print("\n ORT: ", i, item.labels['de'])
            else:
                print("\n ORT: ", i, item.labels['en'])
            check_for_country(item)
    except:
        print("none")
        
#with open ("C:\\Users\\p70076654\\Desktop\\Wiki-test.csv", "w", newline='', encoding="utf-8") as f:
 #   writer = csv.writer(f)
  #  writer.writerows(zip(id_list, wiki_list))
    
print("Done!")
```

```python
# STEP 1: find Wikidata IDs for known Geonames IDs

# open csv file with entity names and IDs

entity_file="https://raw.githubusercontent.com/ieg-dhr/DigiKAR/main/OntologyFiles/Places_Geonames.csv"
df = pd.read_csv(entity_file, index_col=0)
id_list=df['id'].tolist() 

# iterate over IDs in list and find information

wiki_list=[]

for i in id_list:
    i_list=[]
    with open('C:\\Users\\p70076654\\Downloads\\SPARQL_mapGeonamesID.rq', 'r') as query_file:
        QUERY = query_file.read()
            
# create generator based on query
# returns an iterator that produces a sequence of values when iterated over
# useful when creating large sequences of values
    try: 
        wikidata_site = pywikibot.Site("wikidata", "wikidata")
        
# start query using string interpolation        
# https://stackoverflow.com/questions/76762185/matching-specific-geonames-ids-with-wikidata-ids-using-pywikibot
        generator = pg.WikidataSPARQLPageGenerator(QUERY % int(i), site=wikidata_site)

    # iterate over generator and print Wikidata ID 
        for item in generator:

            wikidata_ID=item
            print(wikidata_ID)
            i_list.append(wikidata_ID)
    except:
        print("No entry in Wikidata.")
        i_list.append("none")
    print(i_list) 
    try:
        wiki_list.append(i_list[0])
    except:
        wiki_list.append("n/a")
        
with open ("C:\\Users\\p70076654\\Desktop\\Geonames_Wikidata_AP3.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(zip(id_list, wiki_list))
    
print("Done!")
```
