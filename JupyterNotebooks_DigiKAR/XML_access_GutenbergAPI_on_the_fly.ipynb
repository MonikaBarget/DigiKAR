```python
## import packages

import pandas_read_xml as pdx
from pandas_read_xml import flatten, fully_flatten, auto_separate_tables
import requests
import urllib.request
import os
from bs4 import BeautifulSoup
import bs4.builder._lxml
from xml.etree.ElementTree import XML, fromstring
import re
```

```python
# call landing page URL and get file overview

gutenberg_url="http://gutenberg-biographics.ub.uni-mainz.de/api/items/persons/"

# function to extract html document from given url
# as suggested on https://www.geeksforgeeks.org/beautifulsoup-scraping-link-from-html/

def getHTMLdocument(url):

    # request for HTML document of given url
    response1 = requests.get(url)

    # response will be provided in JSON format
    return response1.text

# Navigate to the application home page

html_document = getHTMLdocument(gutenberg_url)
soup = BeautifulSoup(html_document, 'xml')
print(soup)
print(type(soup))
```

```python
# Find links to XML files by keyword

keywords=["Zwick"] # enter person name

resources = soup.find_all("resource")

for r in resources:
    #print(r)
    for keyword in keywords:
          r_value=r.get("value")
          if keyword in r_value:
              items = soup.find_all('resource', attrs={'value': keyword})
              for i in items:
                  print(i)
                  l = i.get("href")

    # use new URL to access individual XML files

              response2 = requests.get(l) # use new URL to access individual XML files
              soup2 = BeautifulSoup(response2.text, 'xml') # read file as XML
              print(soup2)

    # Write XML to new file on local DRIVE

              response2 = requests.get(l)
              f_xml=directory+"ProfXML/"+str(keyword)+".xml" # Change folder name as you please but do not forgot to create this folder on your DRIVE!
              open(f_xml, 'wb').write(response2.content)

          else:
            #print("Keyword not found!")
            continue

print("Done")
```
