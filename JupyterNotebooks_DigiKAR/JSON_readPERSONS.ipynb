```python
in_directory="JSONdata_bishops"
out_directory="output"
```

```python
items=(os.listdir(in_directory))
for i in items:
    filename=os.path.join(in_directory, i) 
    print(filename)
    try:
        indata=open(filename,"r", encoding="utf-8", errors="ignore")     
        jsondata = json.loads(indata.read())
        #print(jsondata)
        pers_data = jsondata['persons']
    except JSONDecodeError as e:
        print(e)
    
    # Convert JSON data to DataFrame
    df = pd.DataFrame(pers_data)

    # Display the DataFrame
    display(df)
```

```python
# Write the updated DataFrame to an Excel file

df.to_excel('BishopsMainz-fromJSON.xlsx', index=True)

# Show OFFICES content

#print(df["offices"].values)
```
