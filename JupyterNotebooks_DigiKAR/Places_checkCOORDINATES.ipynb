```python
# import packages 
import pandas as pd
import numpy as np
from eldar import Query # for future implementation of BOOLEAN QUERIES
from collections import Counter
import xlsxwriter
import openpyxl

### adjust the file path according to your own file structure

filename = "C:\\Users\\p70076654\\Downloads\\Ortsontologie_Geocoded_geprüft.xlsx"

### read excel file to dataframe

df = pd.read_excel(filename, dtype=str) # axis=1, sort=False
df = df.fillna("@") # replace empty fields for string
display(df)
```

```python
### read factoids from selected column, incl. duplicate values

place_f=(df[['place_name']]) # retrieve data from selected column

#display(place_f) # option to show selected data

place_list=place_f.values.tolist() # convert data frame to sorted list
place_list_flat=[item for sublist in place_list for item in sublist] # flatten list
place_unique=place_f.drop_duplicates() # remove duplicates

#display(place_unique) # option to show selected data excluding duplicates
place_unique_list=place_unique.values.tolist() # write unique values to list
    
print("\n\nYour factoid list contains", len(place_f), "ENTRIES and", len(place_unique), "UNIQUE PLACES.") # count data in selected column
 
```

```python
# iterature through list of values defined above to identify inconsistent mapping across columns

problem_list = []

for p in place_unique_list:
    p_str = p[0]
    geodata_df = df[df['place_name'] == p_str]
    
    # Check if dataframe is valid
    if geodata_df.empty:
        print("No geodata found for", p_str, "!")
    else:
        if all(geodata_df["place_lat_geonames"] == geodata_df["place_lat_geonames"].iloc[0]) and all(geodata_df["place_lng_geonames"] == geodata_df["place_lng_geonames"].iloc[0]):
           print("No divergent geodata for", p, "!")
        else:
           problem_df = geodata_df["place_lat_geonames"]
           print("Divergent coordinates found for", p, ":")
           display(problem_df)
           problem_list.append(p)

print("List of places with inconsistent geodata:", problem_list)
```
