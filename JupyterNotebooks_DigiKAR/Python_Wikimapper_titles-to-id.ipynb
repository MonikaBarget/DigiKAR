```python
# taken from https://pypi.org/project/wikimapper/

from wikimapper import WikiMapper
import os.path
import csv
import sqlite3

# Map Wikidata id to Wikipedia page title

# Define list of territories

infile=("C:\\Users\\p70076654\\Desktop\\Territories_HRE.txt")

title_list=[]
file = open(infile, "rt", encoding="utf-8", errors="ignore")
entities = file.readlines()
for e in entities:
    title_list.append(e[:-1])
    
print(title_list)

# Define list of alternative names

religious_entities=["Fürstpropstei", 
                    "Fürstabtei", 
                    "Propstei", 
                    "Abtei", 
                    "Stift",
                    "Frauenstift",
                    "Damenstift",
                    "Chorherrenstift",
                    "Kloster", 
                    "Gefürstete_Abtei",
                    "Gefürstete_Probstei",
                    "Gefürstete Abtei",
                    "Gefürstete Probstei"]
    
# path for database file

BASE_DIR="C:\\Users\\p70076654\\OneDrive"
db_path = os.path.join(BASE_DIR, "index_dewiki-20190420.db") # new version including German data?

# read titles

titles_result=[]
for title in title_list:
    t_new=title.replace(" ", "_")
    print("Territorium: ", t_new)
    mapper = WikiMapper(db_path)
    wikipedia_id1 = mapper.title_to_id(t_new)
    if len(str(wikipedia_id1))>4:
        print("ID des Territoriums: ", wikipedia_id1)
        titles_result.append(str(wikipedia_id1))
    else:
        t_split=t_new.split("_")
        try:
            t_merged=(str(t_split[-2]) + "_" + str(t_split[-1]))
            print("Alternative Bezeichung :", t_merged)
        except IndexError:
            t_merged=t_new
        mapper = WikiMapper(db_path)
        wikipedia_id2 = mapper.title_to_id(str(t_merged))
        if len(str(wikipedia_id2))>4:
            print("ID des Territoriums: ", wikipedia_id2)
            titles_result.append(str(wikipedia_id2))
        else:
            if "Herrschaft" in t_split:
                t1=(str(t_split[-1]) + "_(Herrschaft)")
                mapper = WikiMapper(db_path)
                wikipedia_id3 = mapper.title_to_id(t1)
                print(wikipedia_id3)
                titles_result.append(str(wikipedia_id3))
            elif "Grafschaft" in t_split:
                t2=t_split[-1]
                mapper = WikiMapper(db_path)
                wikipedia_id4 = mapper.title_to_id(t2)
                print(wikipedia_id4)
                titles_result.append(str(wikipedia_id4))   
            elif t_split[0] in religious_entities:
                rel_list=[]
                for r in religious_entities:
                    t3=(str(r) + "_" + str(t_split[-1]))
                    mapper = WikiMapper(db_path)
                    wikipedia_id5 = mapper.title_to_id(str(t3))
                    if len(str(wikipedia_id5))>4:
                        print(wikipedia_id5)
                        rel_list.append(str(wikipedia_id5))
                print(rel_list)
                try:
                    titles_result.append(rel_list[0]) 
                except IndexError:
                    titles_result.append("None")
            else:
                titles_result.append("None")
                
rows = zip(title_list, titles_result)
                
with open ("C:\\Users\\p70076654\\Desktop\\ReichsunmittelbareTerritorien_Wiki_v5.csv", "w", newline='', encoding="ANSI") as f:
    writer=csv.writer(f, dialect='excel')
    for row in rows:
        writer.writerow(row)
    
print("Done!")
```
